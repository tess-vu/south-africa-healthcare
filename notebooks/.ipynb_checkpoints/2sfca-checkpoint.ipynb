```python
import geopandas as gpd
import pandas as pd
sal= pd.read_csv('../pop_pred_final.csv')
phm=pd.read_csv('../../PHARMACIES_places_full_results.csv')
```

```python
salpoint=gpd.read_file('../../2011_census/2011_Census/2011_ea_sal_centroids_kzn_gp.dbf')
salpoly=gpd.read_file('../../2011_census/2011_Census/ea_sal_kzn_gp.dbf')
```

```python
pharm_gdf = gpd.GeoDataFrame(
    phm,
    geometry=gpd.points_from_xy(phm['lng'], phm['lat']),
    crs="EPSG:4326"  # WGS84
)
pharm_gdf = pharm_gdf.to_crs(salpoly.crs)
```

```python
pharm_with_sal = gpd.sjoin(
    pharm_gdf,
    salpoly[['EA_CODE', 'EA_GTYPE', 'geometry']],
    how='left',
    predicate='within'
)
```

```python
def assign_catchment(row):
    if row['EA_GTYPE'] == 'Urban':
        return 2 * 1609.34
    elif row['EA_GTYPE'] == 'Traditional':
        return 5 * 1609.34
    else:
        return 15 *1609.34

sal['catchment'] = sal.apply(assign_catchment, axis=1)
pharm_with_sal['catchment'] = pharm_with_sal.apply(assign_catchment, axis=1)
```

```python
import osmnx as ox

# bounds = [minx, miny, maxx, maxy]
minx, miny, maxx, maxy = salpoly.total_bounds

bbox = (miny, maxy, minx, maxx)  # (south, north, west, east)

G = ox.graph_from_bbox(
    bbox,
    network_type='drive'
)
```

```python
ox.settings.requests_kwargs = {"verify": False}
```

```python
sal['node'] = ox.nearest_nodes(G, salpoly.geometry.x, salpoly.geometry.y)
pharm_with_sal['node'] = ox.nearest_nodes(G, pharm_with_sal.geometry.x, pharm_with_sal.geometry.y)
```

```python
from tqdm import tqdm
###Finding all population within catchment distance of sal type
Rj_list = []

for idx, pharm in tqdm(pharm_gdf.iterrows(), total=len(pharm_gdf)):
    lengths = nx.single_source_dijkstra_path_length(
        G,
        pharm['node'],
        cutoff=pharm['catchment'],
        weight='length'
    )
    
    reachable_nodes = set(lengths.keys())
    
    # find population points whose nodes are reachable
    nearby_pop = pop_gdf[sal['node'].isin(reachable_nodes)]
    
    total_pop = nearby_pop['sal2023_est'].sum()

    ##over 1 because the higher the pop around a pharm. the weaker the supply 
 
    if total_pop > 0:
        Rj = 1 / total_pop
    else:
        Rj = 0
    
    Rj_list.append(Rj)

pharm_gdf['Rj'] = Rj_list
```

```python
Ai_list = []

###measuring the # of pharmacies within catchment node of SAL 

for idx, pop in tqdm(sal.iterrows(), total=len(salpoint)):
    lengths = nx.single_source_dijkstra_path_length(
        G,
        pop['node'],
        cutoff=pop['catchment'],
        weight='length'
    )
    
    reachable_nodes = set(lengths.keys())
    
    nearby_pharm = pharm_gdf[pharm_gdf['node'].isin(reachable_nodes)]

    
###now were summing all supply ratios of sal node to measure demand    
  
    Ai = nearby_pharm['Rj'].sum()
    
    Ai_list.append(Ai)

sal['Ai'] = Ai_list
```
