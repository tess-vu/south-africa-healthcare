```python
# Kwazulu Natal Walk Analysis 
# Walk Catchment: 2km (2000m) (1.25mi)
# Pharmacy Population Demand Catchment: 8km (8000m) (5mi)

# Pharmacy Supply: 5000/catchment pop (we must assume equal supply accross pharamcies due to limited data)

#Pharmacy Catchment Population Distribution
# mean      681.922009
# std       585.490022
# min         0.000000
# 25%       398.265719
# 50%       642.756608
# 75%       819.792569
# max      6519.646647
```

```python
import osmnx as ox
import networkx as nx
from tqdm import tqdm


kznsal = kznsal.to_crs(kznwalk.graph["crs"])
kznpharm = kznpharm.to_crs(kznwalk.graph["crs"])
```

```python

# ----------------------------
# GRAPH TRUNCATION
(maxy, miny, maxx, minx) = kznsal.total_bounds 
bbox = (maxy, miny, maxx, minx)

kznwalk = ox.truncate.truncate_graph_bbox(kznwalk, bbox)
```

```python

# NODE ASSIGNMENT
# ----------------------------
kznsal["node"] = ox.nearest_nodes(
    kznwalk,
    kznsal.geometry.x,
    kznsal.geometry.y
)

pharm_gdf = kznpharm.copy()
pharm_gdf = pharm_gdf[
    pharm_gdf.geometry.x.notna() &
    pharm_gdf.geometry.y.notna()
]

pharm_gdf["node"] = ox.nearest_nodes(
    kznwalk,
    pharm_gdf.geometry.x,
    pharm_gdf.geometry.y
)

##Distance Decay ---this only needs to be defined once for the whole notebook
import numpy as np

def decay(d, beta=0.001):
    return np.exp(-beta * d)
```

```python
Rj_list = []

for idx, pharm in tqdm(pharm_gdf.iterrows(), total=len(pharm_gdf)):

    lengths = nx.single_source_dijkstra_path_length(
        kznwalk,
        pharm["node"],
        cutoff=3000,
        weight="length"
    )

    weighted_pop = 0

    for node, dist in lengths.items():
        pop = kznsal.loc[kznsal["node"] == node, "sal2023_est"]

        if len(pop) > 0:
            weighted_pop += pop.values[0] * decay(dist)

    Rj = 1 / (weighted_pop / 1000) if weighted_pop > 0 else 0

    Rj_list.append(Rj)

pharm_gdf["Rj"] = Rj_list



Ai_list = []

for idx, pop in tqdm(kznsal.iterrows(), total=len(kznsal)):

    lengths = nx.single_source_dijkstra_path_length(
        kznwalk,
        pop["node"],
        cutoff=3000, #1.86 miles
        weight="length"
    )

    Ai = 0

    for node, dist in lengths.items():
        pharm_subset = pharm_gdf[pharm_gdf["node"] == node]

        if len(pharm_subset) > 0:
            Ai += pharm_subset["Rj"].sum() * decay(dist)

    Ai_list.append(Ai)

kznsal["Ai_walk"] = Ai_list
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# --- QUANTILES ---
allaccess["walk_q"] = pd.qcut(
    allaccess["AI_WALK"].rank(method="first"),
    3,
    labels=[0,1,2]
)

allaccess["drive_q"] = pd.qcut(
    allaccess["AI_DRIVE"].rank(method="first"),
    3,
    labels=[0,1,2]
)

# --- COLORS ---
biv_colors = {
    (0,0): "#f0f0f0",  (0,1): "#c6dbef",  (0,2): "#002395",
    (1,0): "#fcbba1",  (1,1): "#bcbddc",  (1,2): "#807dba",
    (2,0): "#DE3831",  (2,1): "#756bb1",  (2,2): "#54278f",
}

# --- ASSIGN COLORS ---
allaccess["biv_color"] = allaccess.apply(
    lambda r: biv_colors[(int(r["walk_q"]), int(r["drive_q"]))],
    axis=1
)

# =========================
# PLOT (ONLY ONCE)
# =========================
fig, ax = plt.subplots(figsize=(10,10))

allaccess.plot(
    color=allaccess["biv_color"],
    ax=ax,
    linewidth=0.1,
    edgecolor="white"
)
ax.scatter(3118870, -3005460, color='black', s=10, zorder=20)
ax.text(3118870 -15000, -3005460+7000, "Johannesburg", fontsize=9, ha='left')
ax.scatter(3142400, -2952500, color='black', s=10, zorder=20)
ax.text(3142400 , -2952500 + 5000, "Pretoria", fontsize=9)
ax.scatter(3562000, -3336000, color='black', s=10, zorder=20)
ax.text(3562000 - 30000, -3336000+10000, "Empangeni", fontsize=9)
ax.scatter(3453460, -3488260, color='black', s=10, zorder=20)
ax.text(3453460 +5000, -3488260, "Durban", fontsize=9, ha='left')
ax.scatter(3391030, -3337800, color='black', s=10, zorder=20)
ax.text(3391030 - 5000, -3337800 - 15000, "Msinga", fontsize=9, ha='left')
ax.set_title("Bivariate Accessibility: Walking vs Driving Accessibility Scores", fontsize=14)
ax.axis("off")

# =========================
# LEGEND (SAME FIGURE)
# =========================
import matplotlib.patches as mpatches

import matplotlib.patches as mpatches
import pandas as pd

# counts (make sure defined)
counts = pd.crosstab(allaccess["walk_q"], allaccess["drive_q"])

legend_ax = fig.add_axes([0.75, 0.15, 0.2, 0.2])
legend_ax.axis("off")

import matplotlib.patches as mpatches

counts = pd.crosstab(allaccess["walk_q"], allaccess["drive_q"])

legend_ax = fig.add_axes([0.75, 0.15, 0.2, 0.2])
legend_ax.axis("off")

for i in range(3):          # walk (0=low bottom → 2=high top)
    for j in range(3):      # drive (0=low left → 2=high right)

        color = biv_colors[(i, j)]

        rect = mpatches.Rectangle(
            (j, i),   # x = drive, y = walk
            1, 1,
            facecolor=color,
            edgecolor="white"
        )
        legend_ax.add_patch(rect)

        value = counts.loc[i, j] if (i in counts.index and j in counts.columns) else 0

        legend_ax.text(
            j + 0.5,
            i + 0.5,
            f"{value}",
            ha="center",
            va="center",
            fontsize=8,
            color="white"
        )

legend_ax.set_xlim(0, 3)
legend_ax.set_ylim(0, 3)
# Axis labels
legend_ax.text(1.5, -0.4, "Driving Access →", ha="center", fontsize=9)
legend_ax.text(-0.4, 1.5, "Walking Access →", va="center", rotation=90, fontsize=9)
plt.savefig("images/bivariate.png", dpi=300, bbox_inches='tight')

plt.show()
```
