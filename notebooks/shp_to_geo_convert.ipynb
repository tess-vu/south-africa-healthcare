```python
import geopandas as gpd

gdf = gpd.read_file("../data/raw/MDB_Wards_2020.shp")

# Snowflake requires EPSG:4326
gdf = gdf.to_crs(epsg=4326)

gdf.to_file("../data/processed/MDB_Wards_2020.geojson", driver="GeoJSON")
```

```python
import json

in_path = "../data/processed/MDB_Wards_2020.geojson"
out_path = "../data/processed/MDB_Wards_2020.geojsonl"

with open(in_path, "r", encoding="utf-8") as f:
    fc = json.load(f)

with open(out_path, "w", encoding="utf-8") as f:
    for feature in fc["features"]:
        f.write(json.dumps(feature) + "\n")

```
